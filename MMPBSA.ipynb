```python
import subprocess
import os
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
os.chdir('/Users/gil_f.hoben/Troponin_md_processed/MD1/run1/free_energy/')
subprocess.call(['bash', 'mmpbsa_parser.sh'])
os.chdir('./Analysis')

complex_gb_coul_MD1 = np.loadtxt('_MMPBSA_complex_gb.coul')
complex_gb_polar_MD1 = np.loadtxt('_MMPBSA_complex_gb.polar')
complex_gb_vdw_MD1 = np.loadtxt('_MMPBSA_complex_gb.vdw')
complex_gb_surf_MD1 = np.loadtxt('_MMPBSA_complex_gb.surf')
ligand_gb_coul_MD1 = np.loadtxt('_MMPBSA_ligand_gb.coul')
ligand_gb_polar_MD1 = np.loadtxt('_MMPBSA_ligand_gb.polar')
ligand_gb_vdw_MD1 = np.loadtxt('_MMPBSA_ligand_gb.vdw')
ligand_gb_surf_MD1 = np.loadtxt('_MMPBSA_ligand_gb.surf')
receptor_gb_coul_MD1 = np.loadtxt('_MMPBSA_receptor_gb.coul')
receptor_gb_polar_MD1 = np.loadtxt('_MMPBSA_receptor_gb.polar')
receptor_gb_vdw_MD1 = np.loadtxt('_MMPBSA_receptor_gb.vdw')
receptor_gb_surf_MD1 = np.loadtxt('_MMPBSA_receptor_gb.surf')

os.chdir('/Users/gil_f.hoben/Troponin_md_processed/MD1/run1/free_energy_25000_27500/')
subprocess.call(['bash', '/Users/gil_f.hoben/MMPBSA_fileparser/mmpbsa_parser.sh'])
os.chdir('./Analysis')

    
complex_gb_coul_MD1 = np.append(complex_gb_coul_MD1, np.loadtxt('_MMPBSA_complex_gb.coul'))
complex_gb_polar_MD1 = np.append(complex_gb_polar_MD1, np.loadtxt('_MMPBSA_complex_gb.polar'))
complex_gb_vdw_MD1 = np.append(complex_gb_vdw_MD1, np.loadtxt('_MMPBSA_complex_gb.vdw'))
complex_gb_surf_MD1 = np.append(complex_gb_surf_MD1, np.loadtxt('_MMPBSA_complex_gb.surf'))
ligand_gb_coul_MD1 = np.append(ligand_gb_coul_MD1, np.loadtxt('_MMPBSA_ligand_gb.coul'))
ligand_gb_polar_MD1 = np.append(ligand_gb_polar_MD1, np.loadtxt('_MMPBSA_ligand_gb.polar'))
ligand_gb_vdw_MD1 = np.append(ligand_gb_vdw_MD1, np.loadtxt('_MMPBSA_ligand_gb.vdw'))
ligand_gb_surf_MD1 = np.append(ligand_gb_surf_MD1, np.loadtxt('_MMPBSA_ligand_gb.surf'))
receptor_gb_coul_MD1 = np.append(receptor_gb_coul_MD1, np.loadtxt('_MMPBSA_receptor_gb.coul'))
receptor_gb_polar_MD1 = np.append(receptor_gb_polar_MD1, np.loadtxt('_MMPBSA_receptor_gb.polar'))
receptor_gb_vdw_MD1 = np.append(receptor_gb_vdw_MD1, np.loadtxt('_MMPBSA_receptor_gb.vdw'))
receptor_gb_surf_MD1 = np.append(receptor_gb_surf_MD1, np.loadtxt('_MMPBSA_receptor_gb.surf'))
    
    
complex_total_MD1 = complex_gb_coul_MD1 + complex_gb_polar_MD1 + complex_gb_vdw_MD1 + complex_gb_surf_MD1
receptor_total_MD1 = receptor_gb_coul_MD1 + receptor_gb_polar_MD1 + receptor_gb_vdw_MD1 + receptor_gb_surf_MD1
ligand_total_MD1 = ligand_gb_coul_MD1 + ligand_gb_polar_MD1 + ligand_gb_vdw_MD1 + ligand_gb_surf_MD1

print 'Average complex Energy: %.2f kcal/mol' % complex_total_MD1.mean()
print 'Average receptor Energy: %.2f kcal/mol' % receptor_total_MD1.mean()
print 'Average ligand Energy: %.2f kcal/mol' % ligand_total_MD1.mean()
print 'Average Delta Total: %.2f kcal/mol' % (complex_total_MD1.mean()-receptor_total_MD1.mean()-ligand_total_MD1.mean())
print 'Standard Deviation of Delta Total: %.2f' %(complex_total_MD1 - receptor_total_MD1 - ligand_total_MD1).std()

plt.figure(figsize=(12,12))
plt.title('MD1 run1 $\Delta$Total Energy - 50 frame rolling average', size = 22)
plt.plot(pd.DataFrame(complex_total_MD1 - receptor_total_MD1 - ligand_total_MD1).rolling(window=50,center=False).mean()[0], label = 'MD1: Average $\Delta$Total = %.2f kcal/mol\nStandard deviation = %.2f' % ((complex_total_MD1 - receptor_total_MD1 - ligand_total_MD1).mean(), (complex_total_MD1 - receptor_total_MD1 - ligand_total_MD1).std()))
plt.ylabel('$\Delta$Total (kcal/mol)', size =15)
plt.xlabel('Frame', size =15)
plt.legend()
```
